# Words vs Sentences vs Characters?
At this stage, we only explore structure.

```python
#Number of words
import nltk
from nltk.tokenize import word_tokenize

truman_1945_words_tokens = word_tokenize(truman_1945_raw)

print(f"Number of Word Tokens: {len(truman_1945_words_tokens)}")

print(f'First 10 Tokens: {truman_1945_words_tokens[:10]}')
```

```python
#Number of Sentences
from nltk.tokenize import sent_tokenize

truman_1945_sents_tokens = sent_tokenize(truman_1945_raw)

print(f"Number of Word Tokens: {len(truman_1945_sents_tokens)}")

print(f'First 10 Tokens: {truman_1945_sents_tokens[:10]}')
```

```python
#Number of Characters
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'.')
truman_1945_chars_tokens = tokenizer.tokenize(truman_1945_raw)

print(f"Number of Word Tokens: {len(truman_1945_chars_tokens)}")

print(f'First 10 Tokens: {truman_1945_chars_tokens[:10]}')
```

# Quick Corpus-Wide Exploration

```python
from nltk.tokenize import word_tokenize, sent_tokenize, RegexpTokenizer

corpus_raw = state_union.raw(state_union.fileids())

#Word Tokenization
corpus_words_tokens = word_tokenize(corpus_raw)

#Character Tokenization
tokenizer = RegexpTokenizer(r'.')
corpus_chars_tokens = tokenizer.tokenize(corpus_raw)

#Sentence Length Frequency
corpus_sents_tokens = sent_tokenize(corpus_raw)
corpus_sents_len = []
for sent in corpus_sents_tokens:
    corpus_sents_len.append(len(sent))
```
